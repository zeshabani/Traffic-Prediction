```python
segment_df = pd.read_csv("segments_corrected.csv")
segment_df = segment_df.drop((['Unnamed: 0']), axis = 1)
all_segments = segment_df['segment_id'].values
```

```python
def find_good_segments(file_name):
  df = pd.read_csv(file_name)
  df = df.drop((['Unnamed: 0']), axis = 1)
  print(file_name)
 
  segment_arr = df['segment_id'].values
  segment_list = list(segment_arr)
  segment_set = set(segment_arr)
 
  print("segment list:" + str(len(segment_list)))
  print("segment set:" + str(len(segment_set)))
  print("            ")
 
  good_segments = []
  j = 0
  for segment in segment_set:
    if abs(segment) in all_segments and segment_list.count(segment) >= 30:
      good_segments += [segment]
    if j%5000 == 0:
      print(j)
    j += 1
 
 
  good_df = pd.DataFrame(good_segments)
  to_save_address = "good_segment_day" + file_name 
  good_df.to_csv(to_save_address)
  print(len(good_df))
  print("_________________________________________________________")
```

```python
day = 21
%cd /content/gdrive/MyDrive/Neshan/day21
for i in range(10, 11):
  file_name = str(day) + "_part" + str(i) + ".csv"
  find_good_segments(file_name)
```
