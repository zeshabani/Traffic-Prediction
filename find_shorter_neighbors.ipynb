```python
df = pd.read_csv('neighbors_1000.csv')
neighbs = df.values
```

```python
segment_inform_df = pd.read_csv("good_segment_information.csv")
segment_inform = segment_inform_df[['segment_id', 'Latitude', 'Longitude']].values
segment_id = list(segment_inform_df[['segment_id']].values.flatten())
```

```python
n = len(neighbs)
print(n)
neighbors_shorter = []
for i in range(n):
  raw = neighbs[i]
  index1 = segment_id.index(raw[0])
  index2 = segment_id.index(raw[1])
  distance = (segment_inform[index1][1] - segment_inform[index2][1])**2 + (segment_inform[index1][2] - segment_inform[index2][2])**2
  distance = math.sqrt(distance)
  if distance <= 750:
    neighbors_shorter += [raw]
  if i%10000 == 0:
    print(i)
```

```python
neighbors_shorter = np.asarray(neighbors_shorter)
```

```python
new_df = pd.DataFrame(neighbors_shorter, columns=['segment', 'neighbor'])
```

```python
new_df.to_csv("neighbors_750.csv")
```
