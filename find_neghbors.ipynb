```python
good_segments_df = pd.read_csv('good_segment_information2.csv')
good_segment_arr = good_segments_df.values
```

```python
# save_df = pd.DataFrame(columns =['neighbor']) 
# save_df.to_csv('neighbors_1000_2.csv')
```

```python
n = len(good_segment_arr)
begpoint = 20000
endpoint = 30000
new_neigbs = []
print(begpoint)
print(endpoint)
for i in range(begpoint, endpoint):
  for j in range(i+1, n):
    if abs(good_segment_arr[i][2]) != abs(good_segment_arr[j][2]):
      distance = (good_segment_arr[i][3] - good_segment_arr[j][3])**2 + (good_segment_arr[i][4] - good_segment_arr[j][4])**2
      distance = math.sqrt(distance)
      if distance <= 1000:
        new_neigbs += [[good_segment_arr[i][2], good_segment_arr[j][2]]]
print("_____________________")
```

```python
from csv import writer
with open('neighbors_1000_2.csv', 'a') as f_object:
  writer_object = writer(f_object)
  writer_object.writerows(new_neigbs)
  f_object.close()
 
print("beg: " + str(begpoint) + "/ end: " + str(endpoint) + " saved")
```
