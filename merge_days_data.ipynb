```python
import numpy as np
import pandas as pd
from csv import writer
```

```python
%cd /content/gdrive/MyDrive/Neshan/day21
savefile = "good_traffic_data_day21.csv"
```

```python
traffic_df = pd.DataFrame(columns =['speed', 'date_time']) 
traffic_df.to_csv(savefile)
for i in range(1, 11):
  print("part" + str(i) + ":")
  good_filename = "good_segment_day21_part" + str(i) + ".csv"
  all_filename = "21_part" + str(i) + ".csv"
 
  good_df = pd.read_csv(good_filename)
  all_df = pd.read_csv(all_filename)
 
  if len(good_df) != 0:
    good_segments = good_df['0'].values
    print("length of good segments: " + str(len(good_segments)))
 
    count = 0
    for segment in good_segments:
      temp = all_df[all_df["segment_id"] == segment]
      temp_var = temp[['segment_id', 'speed', 'date_time']].values
 
      with open(savefile, 'a') as f_object:
        writer_object = writer(f_object)
        writer_object.writerows(temp_var)
        f_object.close()
      if count % 1000 == 0:
        print(count)
      count += 1
  else:
      print("length of good_df: " + str(0))
  print("___________________________")
```
