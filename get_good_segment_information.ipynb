```python
import pandas as pd
import numpy as np
import csv
from csv import writer
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
%cd /content/gdrive/MyDrive/Neshan/good_traffic_data/
```

```python
all_segments = []
for i in range(1, 22):
  if i != 7 and i != 14:
    filename = "good_traffic_data_day" + str(i) + ".csv"
    good_df = pd.read_csv(filename)
    good_arr = good_df['Unnamed: 0'].values
    good_set = set(good_arr)
    good_list = list(good_set)
    all_segments += good_list.copy()
```

```python
all_segments_set = set(all_segments)
all_segments = list(all_segments_set)
```

```python
%cd /content/gdrive/MyDrive/Neshan/
df = pd.read_csv('segments_corrected.csv')
df = df.drop((['Unnamed: 0']), axis = 1)
```

```python
save_df = pd.DataFrame(columns =['name', 'segment_id', 'Latitude', 'Longitude']) 
save_df.to_csv('good_segment_information2.csv')
```

```python
print(len(all_segments))
count = 0
for segment in all_segments:
  temp = df[df["segment_id"] == abs(segment)]
  temp_var = temp.values
  temp_var[0][2] = segment  
  with open('good_segment_information2.csv', 'a') as f_object:
    writer_object = writer(f_object)
    writer_object.writerows(temp_var)
    f_object.close()
  if count % 1000 == 0:
    print(count)
  count += 1
```
